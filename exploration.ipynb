```python
# imports

import numpy as np
import matplotlib as plt
import pandas as pd
import csv
```

```python
#import data
train_data = []
columns = []
with open('train.csv', newline='') as csvfile:
    train_data_reader = csv.reader(csvfile)
    first = True
    for line in train_data_reader:
        if first:
            columns.append(line)
            first = False
        else:
            train_data.append(line)
train_data = pd.DataFrame(train_data, columns=columns)
```

```python
test_data = []
with open('test.csv', newline='') as csvfile:
    test_data_reader = csv.reader(csvfile)
    for line in test_data_reader:
        test_data.append(line)
test_data = pd.DataFrame(test_data)
```

```python
num_data_train = []
for col in train_data.columns:
    try:
        train_data[col] = train_data[col].replace(',', '').astype(float)
    except:
        pass
    if train_data[col].dtype in ['float64', 'int64']:
        num_data_train.append(col)

print(num_data_train)
correlation_matrix = train_data[num_data_train].corr()
corr_saleprice = correlation_matrix['SalePrice']
print()
```
